# Plots


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cf_comparison_percentages.csv')
df.head(2)
```

```python
df_changes = df.iloc[:, 0:16]

plt.figure(figsize=(12, 8))
ax = sns.heatmap(
    df_changes,
    cmap="RdBu_r",
    center=0,
    annot=True,
    fmt=".1f",
    cbar_kws={'label': 'Percentage of change (%)'}
)

# Cambiar etiquetas del eje y de 1 a 10
ax.set_yticks([i + 0.5 for i in range(len(df_changes))])
ax.set_yticklabels([str(i + 1) for i in range(len(df_changes))], rotation=0)

plt.xlabel("Task durations and costs")
plt.ylabel("Counterfactuals")
plt.tight_layout()

plt.savefig("cf_changes_heatmap.png", dpi=300)
plt.show()
```
